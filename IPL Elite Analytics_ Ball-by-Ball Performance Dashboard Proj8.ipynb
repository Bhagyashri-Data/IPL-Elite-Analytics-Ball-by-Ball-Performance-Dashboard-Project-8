```python
#Task 3 (Team Performance Analysis): Calculate the number of matches played and won by each team across different cities.
```

```python
city_analysis = df.groupby(['city', 'batting_team']).size().reset_index(name='matches_count')
```

```python
city_analysis.head(10)
```

```python
#Task 4 (Chasing & Target Analytics): Analyze which teams successfully chased down the highest target runs.
```

```python
chasing_data = df[df['result'] == 1]
top_chases = chasing_data.groupby('batting_team')['target_runs'].max().reset_index(name='highest_chase')
```

```python
top_chases = top_chases.sort_values(by='highest_chase', ascending=False)
top_chases
```

```python
#Task 5 (Run Rate & Spike Detection): Track how the run rate changes across different overs (1 to 20) and detect any scoring spikes.
```

```python
over_analysis = df.groupby('balls_left')['cur_run_rate'].mean().reset_index()
```
